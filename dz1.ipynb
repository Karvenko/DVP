```python
import numpy as np
import pandas as pd

from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
import seaborn as sns

from pylab import rcParams
rcParams['figure.figsize'] = 9,9
```

## Задача - предсказать точку на круге в зависимости от X.

```python
n_samples = 1000

randoms = np.random.random(n_samples)
#Making phi [0, 2*PI]
phi = 2 * np.pi * randoms 

#Making x & y
x = np.cos(phi)
y = np.sin(phi)
```

```python
model = LinearRegression()
result = model.fit(np.array(x).reshape(-1, 1), y)

print(model.intercept_, model.coef_)
```

```python
x_lin = np.linspace(-1, 1)
y_lin = model.coef_*x_lin + model.intercept_
plt.plot(x_lin, y_lin, 'r')

plt.scatter(x, y)
plt.show()
```

Линейная регрессия нашла прямую у = 0 (весьма близко к ней). Но любая другая прямая, проходящая через центр имеет такую же ошибку

```python
#Let's square!
x_2 = x**2
y_2 = y**2

model = LinearRegression()
result = model.fit(np.array(x_2).reshape(-1, 1), y_2)

print(model.intercept_, model.coef_)
```

```python
x_lin = np.linspace(-1, 1)
y_lin = model.coef_*x_lin + model.intercept_
plt.plot(x_lin, y_lin, 'r')

plt.scatter(x_2, y_2)
plt.show()
```

Получилась правильная прямая - y^2 = 1. Данная прямая позволит правильно предсказывать модуль у, так как каждому значению х может соответствовать 2 значения у.
